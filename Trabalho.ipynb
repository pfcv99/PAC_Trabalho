```python
df = pd.read_csv('data/raw_data/Mental_health_Depression_disorder_Data.csv', index_col=0)
df.head()
```

```python
df_short = df[df["Code"]=="PRT"]
# Example: Drop rows with NaN values in specific columns
df_short
```

```python
import plotly.express as px

# Assuming you have a DataFrame named df_short with columns "Year", "Schizophrenia (%)", "Entity", and "Code"
# Make sure to replace df_short with the actual name of your DataFrame

fig = px.scatter(df_short, x="Year", y="Schizophrenia (%)", animation_frame="Year", animation_group="Entity",
                size="Schizophrenia (%)", color="Code", hover_name="Entity",
                log_x=True, size_max=55, range_x=[1990, 2020], range_y=[0, 1])

# Customize layout
fig.update_layout(
    title="Schizophrenia (%) Over Years",
    xaxis_title="Year",
    yaxis_title="Schizophrenia (%)",
    legend_title="Country Code",
    font=dict(family="Arial", size=12)
)

# Show the plot
fig.show()
```

```python
def grafico1():
    df = pd.read_csv('data/raw_data/Mental_health_Depression_disorder_Data.csv', index_col=0)
    fig = px.scatter(df, x="Anxiety disorders (%)", y="Depression (%)", animation_frame="Year", animation_group="Entity",
               size="Anxiety disorders (%)", color="Code", hover_name="Entity",
               log_x=True, size_max=55, range_x=[100,100000], range_y=[25,90])

    fig["layout"].pop("updatemenus") # optional, drop animation buttons
    fig.show()

grafico1()
    
```

```python
def main():
    while True:
        print('0 - Terminar\n1 - Obter DataFrame\n2 - Gráficos\n3 - Tabelas\n4 - x\n')
        opt1 = input('Insira uma opção: ')
        if opt1 == "0":
            break
        elif opt1 == "1":
            print('Opção 1')
        elif opt1 == "2":
            print('Gráficos')
            while True:
                print('0 - Voltar\n1 - Gráfico 1\n2 - Gráfico 2\n3 - Gráfico 3\n')
                opt2 = input('Insira uma opção: ')
                if opt2 == '0':
                    break
                elif opt2 == '1':
                    print('gráfico 1')
                    grafico1()
                elif opt2 == '2':
                    print('gráfico 2')
                elif opt2 == '3':
                    print('gráfico 3')
                else:
                    print('Opção inválida. A opção deve estar entre 0 e 3.')
                 
        elif opt1 == "3":
            print('Opção 3')       
        elif opt1 == "4":
            print('Opção 4')      
        else:
            print('Opção inválida. A opção deve estar entre 0 e 4.')
    
main()
```
